#### Fase 4: Evaluación de Diferencias en Reservas de Vuelos por Nivel Educativo
Utilizando un conjunto de datos que hemos compartido, se busca evaluar si existen diferencias significativas en el número de vuelos reservados según el nivel educativo de los clientes

```python
# ¿Existen diferencias significativas en el numero de vuelos reservados segun el nivel educativo de los clientes?
# 1ª preparación de datos (filtrar columnas relevantes como 'Flights Booked' y 'Education')
# 2ª hacer un analisis descriptivo:
#   - agrupar los datos por nivel educativo
#   - calcular la estadistica descriptiva basica como promedio y deviacion del numero de vuelos reservados para cada grupo.

df_4 = df_final[['Education', 'Flights Booked']]

df_4.groupby('Education')['Flights Booked'].agg(['mean', 'std'])
```

```python
plt.figure(figsize=(8, 5))
sns.lineplot(
    data=df_final,
    x='Education',
    y='Flights Booked',
    marker='o',
    errorbar='sd'          # Banda sombreada = desviación estándar de los datos
)
plt.xlabel('Nivel de Educación')
plt.ylabel('Vuelos Reservados')
plt.show()
```

- El nivel educativo no parece influir de forma importante en cuántos vuelos reserva una persona en promedio
